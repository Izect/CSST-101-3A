Question: If the probability of event A is 0.3 and the probability of event B is 0.4, what is the joint probability of A and B assuming they are independent?

```python
def joint_probability(prob_A, prob_B):
    return prob_A * prob_B

prob_A = 0.3
prob_B = 0.4

joint_prob = joint_probability(prob_A, prob_B)

correct_joint_prob = 0.3 * 0.4

if joint_prob == correct_joint_prob:
    print(f"Correct! Joint Probability of A and B: {joint_prob}")
    print("The joint probability of two independent events A and B is the product of their individual probabilities: P(A ∩ B) = P(A) × P(B).")
else:
    print(f"Incorrect. Your result was {joint_prob}. Hint: Remember that for independent events, joint probability is P(A) * P(B).")

```

Exercise 2: Decision-Making Under Uncertainty

Simulate a Decision Process

```python
import random

prob_success = 0.7
prob_failure = 1 - prob_success


reward = 100
cost = -50

expected_value = (prob_success * reward) + (prob_failure * cost)

print(f"Expected Value of making the decision: {expected_value}")
```

Decision Tree

```python
move_forward_prob_success = 0.6
move_forward_reward = 150
move_forward_cost = -100

turn_left_prob_success = 0.8
turn_left_reward = 80
turn_left_cost = -30


ev_move_forward = (move_forward_prob_success * move_forward_reward) + \
                  ((1 - move_forward_prob_success) * move_forward_cost)

ev_turn_left = (turn_left_prob_success * turn_left_reward) + \
               ((1 - turn_left_prob_success) * turn_left_cost)

best_decision = "Move forward" if ev_move_forward > ev_turn_left else "Turn left"

print(f"Expected Value for moving forward: {ev_move_forward}")
print(f"Expected Value for turning left: {ev_turn_left}")
print(f"The best decision is: {best_decision}")
```

Assessment Task 2: Bayesian Inference

```python
def bayesian_inference(prior, likelihood, evidence):
    posterior = (likelihood * prior) / evidence
    return posterior

prior_disease = 0.01
accuracy_positive = 0.9
false_positive_rate = 0.05

prob_positive = (accuracy_positive * prior_disease) + (false_positive_rate * (1 - prior_disease))

likelihood_positive = accuracy_positive
evidence_positive = prob_positive

probability_disease_given_positive = bayesian_inference(prior_disease, likelihood_positive, evidence_positive)

result = probability_disease_given_positive
result
```

Exercise 3: Applying Probability Theory in AI for Diagnosis

```python
def calculate_posterior_probability(prior_disease_prob, sensitivity, false_positive_rate, prior_no_disease_prob):
    total_probability_positive_test = (sensitivity * prior_disease_prob) + (false_positive_rate * prior_no_disease_prob)
    posterior_probability_disease_given_positive = (sensitivity * prior_disease_prob) / total_probability_positive_test
    return posterior_probability_disease_given_positive

prior_disease_probability = 0.02
sensitivity_of_test = 0.95
false_positive_rate_of_test = 0.1
prior_no_disease_probability = 1 - prior_disease_probability

posterior_probability = calculate_posterior_probability(prior_disease_probability, sensitivity_of_test, false_positive_rate_of_test, prior_no_disease_probability)

print(f"The probability of having the disease given a positive test result is: {posterior_probability:.4f}")

```

Assessment Task 3: Decision-Making Under Uncertainty

```python
import numpy as np

def decision_simulation(num_iterations, prob_success, reward, penalty):
    outcomes = np.random.choice([reward, penalty], size=num_iterations, p=[prob_success, 1 - prob_success])
    return outcomes.sum()

num_iterations = 1000
probability_of_success = 0.7
reward = 1000
penalty = -500

simulation_results = decision_simulation(num_iterations, probability_of_success, reward, penalty)

average_return = simulation_results / num_iterations

print(f"Total Return after {num_iterations} iterations: ${simulation_results}")
print(f"Average Return per iteration: ${average_return:.2f}")

def analyze_effect_of_probability(prob_success):
    outcomes = decision_simulation(num_iterations, prob_success, reward, penalty)
    average_return = outcomes / num_iterations
    return average_return

probabilities = [0.5, 0.6, 0.7, 0.8, 0.9]
results = {p: analyze_effect_of_probability(p) for p in probabilities}

print("\nAverage Returns for Different Probabilities of Success:")
for prob, avg_return in results.items():
    print(f"Probability of Success: {prob:.1f}, Average Return: ${avg_return:.2f}")
```

Exercise 4: Probability Distribution in AI

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

n = 1000  # Number of samples
p = 0.5   # Probability of success
mu = 0    # Mean
sigma = 1  # Standard deviation

binomial_data = np.random.binomial(n=1, p=p, size=n)

normal_data = np.random.normal(loc=mu, scale=sigma, size=n)

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].hist(binomial_data, bins=2, color='blue', alpha=0.7)
axs[0].set_title('Binomial Distribution (p=0.5)')
axs[0].set_xticks([0, 1])
axs[0].set_xticklabels(['Failure', 'Success'])
axs[0].set_ylabel('Frequency')
axs[0].set_xlabel('Outcomes')

axs[1].hist(normal_data, bins=30, color='orange', alpha=0.7)
axs[1].set_title('Normal Distribution (μ=0, σ=1)')
axs[1].set_ylabel('Frequency')
axs[1].set_xlabel('Value')

plt.tight_layout()
plt.show()
```

Assessment Task 4: Real-World Scenario

```python
import numpy as np
import matplotlib.pyplot as plt

def conditional_probability_rain_given_evidence(p_rain, p_humidity_given_rain, p_cloud_cover_given_rain, p_humidity, p_cloud_cover):
    p_evidence_given_rain = p_humidity_given_rain * p_cloud_cover_given_rain
    p_evidence = p_humidity * p_cloud_cover
    p_rain_given_evidence = (p_evidence_given_rain * p_rain) / p_evidence
    return p_rain_given_evidence

def plot_prediction(humidity_values, cloud_cover_values, probabilities):
    plt.figure(figsize=(10, 6))
    plt.plot(humidity_values, probabilities, label='Rain Probability based on Humidity')
    plt.plot(cloud_cover_values, probabilities, label='Rain Probability based on Cloud Cover', linestyle='--')
    plt.xlabel('Humidity / Cloud Cover (%)')
    plt.ylabel('Probability of Rain')
    plt.title('Probability of Rain based on Evidence (Humidity and Cloud Cover)')
    plt.legend()
    plt.grid(True)
    plt.show()

p_rain = 0.3
p_humidity_given_rain = 0.8
p_cloud_cover_given_rain = 0.9
p_humidity = 0.6
p_cloud_cover = 0.7

humidity_values = np.linspace(0, 1, 100)
cloud_cover_values = np.linspace(0, 1, 100)

probabilities = [conditional_probability_rain_given_evidence(p_rain, h, c, p_humidity, p_cloud_cover) for h, c in zip(humidity_values, cloud_cover_values)]

plot_prediction(humidity_values, cloud_cover_values, probabilities)
```

Exercise 5: Real-World Application of Probability in AI

```python
def conditional_probability_purchase_given_cart(p_purchase, p_add_to_cart_given_purchase, p_add_to_cart):
    p_purchase_given_cart = (p_add_to_cart_given_purchase * p_purchase) / p_add_to_cart
    return p_purchase_given_cart

p_purchase = 0.2
p_add_to_cart_given_purchase = 0.8
p_add_to_cart = 0.5

purchase_probability = conditional_probability_purchase_given_cart(p_purchase, p_add_to_cart_given_purchase, p_add_to_cart)

print(f"Probability of making a purchase given that an item is added to the cart: {purchase_probability:.2f}")
```

Assessment Task 5: Probability Distributions Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm

def plot_binomial(n, p):
    x = np.arange(0, n + 1)
    pmf = binom.pmf(x, n, p)
    plt.bar(x, pmf, label=f'Binomial Distribution (n={n}, p={p})', alpha=0.7)
    plt.title('Binomial Distribution')
    plt.xlabel('Number of Successes (Heads)')
    plt.ylabel('Probability')
    plt.legend()
    plt.grid()
    plt.show()

def plot_normal(mu, sigma):
    x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
    pdf = norm.pdf(x, mu, sigma)
    plt.plot(x, pdf, label=f'Normal Distribution (μ={mu}, σ={sigma})')
    plt.title('Normal Distribution')
    plt.xlabel('Value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid()
    plt.show()

def simulate_coin_flips(n_trials, p_heads):
    coin_flips = np.random.binomial(n_trials, p_heads)

    outcomes = np.arange(0, n_trials + 1)
    probabilities = binom.pmf(outcomes, n_trials, p_heads)

    plt.bar(outcomes, probabilities, label='Binomial Distribution (1000 trials)', alpha=0.7)
    plt.title(f'Simulation of Coin Flips: {n_trials} Trials')
    plt.xlabel('Number of Heads')
    plt.ylabel('Probability')
    plt.xlim(0, n_trials)
    plt.legend()
    plt.grid()
    plt.show()

n = 10
p = 0.5
mu = n * p
sigma = np.sqrt(n * p * (1 - p))

# Plotting the distributions
plot_binomial(n, p)
plot_normal(mu, sigma)

# Simulate coin flips
n_trials = 1000  # number of coin flips
simulate_coin_flips(n_trials, p)
```
