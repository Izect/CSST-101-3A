Part 1: Implement Basic Probability Calculations

```python
def joint_probability(p_a, p_b):
    return p_a * p_b

def marginal_probability(p_joint, p_b):
    return p_joint / p_b

def conditional_probability(p_a, p_b_given_a):
    return (p_b_given_a * p_a) / (p_b_given_a * p_a + (1 - p_a) * (1 - p_b_given_a))
0
def main():
    print("Welcome to the Probability Calculator!")

    p_a = float(input("Enter the probability of event A (0 to 1): "))
    p_b = float(input("Enter the probability of event B (0 to 1): "))

    joint_prob = joint_probability(p_a, p_b)
    print(f"Joint Probability P(A and B): {joint_prob:.4f}")

    p_joint = joint_prob
    marginal_prob = marginal_probability(p_joint, p_b)
    print(f"Marginal Probability P(A): {marginal_prob:.4f}")

    p_b_given_a = float(input("Enter the conditional probability P(B|A) (0 to 1): "))
    conditional_prob = conditional_probability(p_a, p_b_given_a)
    print(f"Conditional Probability P(A|B): {conditional_prob:.4f}")

if __name__ == "__main__":
    main()
```

Part 2: Bayesian Inference for Real-World Scenarios


```python
def bayesian_inference(prior, likelihood, p_positive):
    posterior = (likelihood * prior) / p_positive
    return posterior

def main():
    print("Welcome to the Bayesian Inference Calculator for Medical Scenarios!")

    prior = float(input("Enter the prior probability of having the disease (0 to 1): "))
    likelihood = float(input("Enter the likelihood of a positive test given the disease (0 to 1): "))
    p_positive = float(input("Enter the overall probability of a positive test (0 to 1): "))

    posterior = bayesian_inference(prior, likelihood, p_positive)

    print(f"The posterior probability of having the disease after a positive test result is: {posterior:.4f}")


print("Running Bayesian Inference Calculator...")
main()
```

Part 3: Simulate Decision-Making Under Uncertainty

```python
import numpy as np

def simulate_decision(prob_success, reward, penalty, trials):
    results = []

    for _ in range(trials):
        if np.random.rand() < prob_success:
            results.append(reward)
        else:
            results.append(-penalty)

    average_return = np.mean(results)
    return average_return
def main():
    print("Welcome to the Decision-Making Simulation!")

    prob_success = float(input("Enter the probability of success (0 to 1): "))
    reward = float(input("Enter the reward for success: "))
    penalty = float(input("Enter the penalty for failure: "))
    trials = int(input("Enter the number of trials to simulate: "))

    average_return = simulate_decision(prob_success, reward, penalty, trials)

    print(f"The average return after {trials} trials is: {average_return:.2f}")

print("Running Decision-Making Simulation...")
main()
```

Part 4: Visualize Probability Distributions

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

def plot_binomial_distribution(n, p):
    trials = np.random.binomial(n, p, 10000)

    plt.hist(trials, bins=n + 1, density=True, alpha=0.6, color='blue')
    plt.title(f'Binomial Distribution (n={n}, p={p})')
    plt.xlabel('Number of Successes')
    plt.ylabel('Probability')
    plt.xticks(np.arange(0, n + 1))
    plt.grid()
    plt.show()

def plot_normal_distribution(mean, std_dev):
    x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
    probabilities = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

    plt.plot(x, probabilities, color='red')
    plt.title(f'Normal Distribution (mean={mean}, std_dev={std_dev})')
    plt.xlabel('Value')
    plt.ylabel('Probability Density')
    plt.grid()
    plt.show()

def interactive_distribution():
    print("Choose a distribution to simulate:")

    distribution_type = widgets.Dropdown(
        options=['Binomial', 'Normal'],
        description='Distribution:',
        disabled=False,
    )

    n_widget = widgets.IntSlider(value=10, min=1, max=100, step=1, description='Trials (n):')
    p_widget = widgets.FloatSlider(value=0.5, min=0.0, max=1.0, step=0.01, description='Success Probability (p):')

    mean_widget = widgets.FloatSlider(value=70.0, min=0.0, max=100.0, step=1.0, description='Mean:')
    std_dev_widget = widgets.FloatSlider(value=15.0, min=1.0, max=30.0, step=1.0, description='Std Dev:')

    def update_plot(change):
        plt.clf()
        if distribution_type.value == 'Binomial':
            plot_binomial_distribution(n_widget.value, p_widget.value)
        else:
            plot_normal_distribution(mean_widget.value, std_dev_widget.value)

    def display_controls(change):
        if change['new'] == 'Binomial':
            display(n_widget)
            display(p_widget)
            n_widget.observe(update_plot)
            p_widget.observe(update_plot)
            mean_widget.close()
            std_dev_widget.close()
        else:
            display(mean_widget)
            display(std_dev_widget)
            mean_widget.observe(update_plot)
            std_dev_widget.observe(update_plot)
            n_widget.close()
            p_widget.close()

    distribution_type.observe(display_controls)

    display(distribution_type)

interactive_distribution()
```

Part 5: Real-World Scenario Prediction

```python
def calculate_probability_of_rain(humidity, cloud_cover):


    base_probability = 0.1
    humidity_factor = 0.3 if humidity > 80 else 0.1
    cloud_cover_factor = 0.4 if cloud_cover > 70 else 0.1

    probability_of_rain = base_probability + humidity_factor + cloud_cover_factor

    return min(probability_of_rain, 1)

def main():
    print("Welcome to the Rain Prediction System!")

    humidity = float(input("Enter the humidity percentage (0-1): "))
    cloud_cover = float(input("Enter the cloud cover percentage (0-1): "))

    if not (0 <= humidity <= 1) or not (0 <= cloud_cover <= 1):
        print("Please enter valid percentages between 0 and 100.")
        return

    probability_of_rain = calculate_probability_of_rain(humidity, cloud_cover)

    print(f"The calculated probability of rain is: {probability_of_rain:.2f}")

    if probability_of_rain > 0.5:
        print("It's likely to rain. Don't forget your umbrella!")
    else:
        print("It's unlikely to rain. Enjoy your day!")

print("Running Rain Prediction System...")
main()
```
