Part 2: Basic Probability Calculation

*   P(Disease) = 0.02 (Prior probability of the disease)
*   P(Positive | Disease) = 0.95 (Sensitivity)
*   P(Positive | NoDisease) = 0.05 (False positive rate)
*   P(Positive) = 0.06 (Probability of a positive test in the population)




CALCULATION

P(Disease | Positive) = P(Positive | Disease) × P(Disease) / P(Positive)

Isert value:
P(Disease | Positive)= 0.95 x 0.02/0.06

0.95 x 0.02 = 0.019

P(Disease | Positive) = 0.019 / 0.06

P(Disease | Positive) = 0.3167



**Findings:**
Posterior Probability: The posterior probability that the patient has heart disease given a positive test result is approximately 0.3167 or 31.67%.

Part 4: Visualizing a Probability Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n = 5
p = 0.5

x = np.arange(0, n + 1)

probabilities = binom.pmf(x, n, p)

plt.figure(figsize=(10, 6))
plt.bar(x, probabilities, color='red', alpha=0.9)
plt.xticks(x)
plt.xlabel('Number of Successes', fontsize=14)
plt.ylabel('Probability', fontsize=14)
plt.title('Binomial Distribution: Probability of Success in 5 Trials', fontsize=16)
plt.ylim(0, max(probabilities) + 0.05)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i in range(len(x)):
    plt.text(x[i], probabilities[i], f'{probabilities[i]:.2f}', ha='center', va='bottom')

plt.show()
```
